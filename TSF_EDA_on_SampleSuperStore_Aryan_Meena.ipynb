```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.pairplot(df, hue = 'Ship Mode')
# Profit and sale have moderate positive correlation other than that we can see no correlation between other attributes
```

```python
df['Ship Mode'].value_counts()

# The most preffered shipping mode is Standard Class with total 5968 shippings
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['Ship Mode'])
# Standard class is most preferred shipping mode followed by Second class
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x = 'Segment',data = df, palette = 'rainbow')

# In Segment Consumer has the highest no. of observation
```

```python
plt.figure(figsize=(6,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%.2f')
plt.show()

# In Sub-Category the "Maximum" dominated sub-category are Binders, Paper, 
# furnishings, Phones, storage, art, accessories and minimum from copiers, machines, suppliers
```

```python
df['State'].value_counts()

# California has the highest number of customers
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='State',data=df,order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()

# Highest number of buyers are from California and New York
```

```python
plt.figure(figsize=(6,4))
df.hist(figsize=(8,8),bins=50)
plt.show()

# Most customers tends to buy quantity of 2 and 3
# Discount give maximum is 0 to 20 percent
```

```python
plt.figure(figsize=(6,4))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()

# Most of the cumstomers are from Western region
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Sales'],df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

# Moderate Positive Correlation between Profit and Sales
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='Discount',y='Profit',label='Profit',color='orange',data=df)
plt.legend()
plt.show()

# No correlation between profit and discount
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['hotpink','green'],figsize=(6,4))
plt.ylabel('Profit/Loss and sales')
plt.show()

# Profit and sales are maximum in consumer segment and minimum in Home Office segment
```

```python
plt.figure(figsize=(6,4))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',
            order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()

# Segment wise sales are almost same in every region
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],
                                                             figsize=(6,4))
plt.ylabel('Profit/Loss and sales')
plt.show()

# Profit and sales are maximum in west region and minimum in south region
```

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',
                                                            ascending=False)
ps[:].plot.bar(color=['green','red'],figsize=(8,4))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()

# Phones sub-category have high sales.
# chairs have high sales but less profit compared to phones
# Tables and Bookmarks sub-categories facing huge loss
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow',
                                              'purple'],alpha=0.9,figsize=(6,4))
plt.ylabel('Profit/Loss and sales')
plt.show()

# Technology and Office Supplies have high profit.
# Furniture have less profit
```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',
                                                                    ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(10,5))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()

# high profit is for california, new york
# loss is for texas, pennsylvania, Ohio
```
